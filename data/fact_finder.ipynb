```python
import wikipediaapi

# Créez une instance de l'API Wikipédia avec la langue souhaitée
wiki = wikipediaapi.Wikipedia("fr")  # Utiliser le français comme langue

# Recherchez une page Wikipédia
page_title = "Marseille"
page = wiki.page(page_title)

# Vérifiez si la page existe et affichez le contenu
if page.exists():
    print(f"Titre de la page : {page.title}")
    print(f"Résumé : {page.summary[0:100]}...")  # Affichez les 100 premiers caractères du résumé
    print(f"Contenu complet : \n{page.text}")
else:
    print(f"La page '{page_title}' n'existe pas.")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://fr.wikipedia.org/wiki/Marseille"

# Effectuer une requête GET pour récupérer le contenu HTML
response = requests.get(url)

# Vérifiez si la requête a réussi
if response.status_code == 200:
    # Analyser le contenu HTML avec BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Récupérer et afficher le titre de la page
    title = soup.find("h1", {"id": "firstHeading"}).text
    print(f"Titre de la page : {title}")

    # Récupérer le contenu principal de la page
    content = soup.find("div", {"id": "bodyContent"})

    # Afficher le contenu texte de chaque paragraphe
    for paragraph in content.find_all("p"):
        print(paragraph.text)
else:
    print(f"Erreur lors de la récupération du contenu de l'URL : {url}")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://fr.wikipedia.org/wiki/Marseille"

# Effectuer une requête GET pour récupérer le contenu HTML
response = requests.get(url)

# Vérifiez si la requête a réussi
if response.status_code == 200:
    # Analyser le contenu HTML avec BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Récupérer et afficher le titre de la page
    title = soup.find("h1", {"id": "firstHeading"}).text
    print(f"Titre de la page : {title}")

    # Trouver la rubrique "Histoire"
    histoire_header = soup.find("span", {"id": "Histoire"})
    if histoire_header:
        histoire_section = histoire_header.find_parent("h2")
        content = histoire_section.find_next_sibling()

        # Afficher le contenu texte de chaque paragraphe de la rubrique "Histoire"
        while content and content.name != "h2":
            if content.name == "p":
                print(content.text)
            content = content.find_next_sibling()
    else:
        print("La rubrique 'Histoire' est introuvable.")

else:
    print(f"Erreur lors de la récupération du contenu de l'URL : {url}")
```

```python
import requests
from bs4 import BeautifulSoup

search_query = "histoire de lyon"  # the query you want to search on Google

# make a request to Google with the search query
url = f"https://www.google.com/search?q={search_query}"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
response = requests.get(url, headers=headers)

# parse the HTML content with BeautifulSoup and get the first search result link
soup = BeautifulSoup(response.content, "html.parser")
search_results = soup.find_all("div", class_="yuRUbf")
first_result = search_results[0].find("a")["href"]

print(first_result)  # print the first search result link
```

```python
import requests
from urllib.parse import unquote

url = "https://fr.wikipedia.org/wiki/Paris"
title = unquote(url.split("/")[-1])
api_url = "https://fr.wikipedia.org/w/api.php"
params = {
    "action": "query",
    "format": "json",
    "prop": "extracts",
    "explaintext": 1,
    "titles": title,
    "exsectionformat": "wiki",
    "exlimit": 1,
    "exintro": 0,
    "generator": "search",
    "gsrsearch": f"intitle:{title} intitle:Histoire",
    "gsrlimit": 1
}
response = requests.get(api_url, params=params)
data = response.json()
histoire_content = None

for page in data["query"]["pages"].values():
    if "extract" in page:
        histoire_content = page["extract"]
        break
```
